```python
#In feature selection we have to remove independent features which are corr [it means independent feature have much corr with independent feature the we can remove , one of it by multicolinearity]
def correlation(dataset,threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
## threshold--Domain expertise
corr_features = correlation(X_train,0.85)
corr_features
```

```python
 X_train.drop(corr_features,axis=1,inplace=True)
```

# Linear Regression

* Mean Squared Error (MSE) is a measure of the average squared differences between predicted values and actual values. It quantifies how well a model's predictions approximate the true data points.

* R-squared (R²) is a statistical measure that tells us how well a regression model explains the variability of the data. Here’s a breakdown:

What it Measures: R² measures how much of the variation in the outcome (or dependent variable) can be explained by the predictors (or independent variables) in the model.

Values:

R² = 0: This means that the model does not explain any of the variation in the data. The predictions are no better than just using the average of the actual values.
R² = 1: This means that the model perfectly explains the variation in the data. The predictions match the actual values exactly.
Values between 0 and 1: The closer the R² value is to 1, the better the model explains the data.
Example:

Imagine you’re trying to predict a student's exam score based on the number of hours they studied. If your model has an R² of 0.8, it means that 80% of the variation in exam scores can be explained by the hours studied. The remaining 20% is due to other factors (like sleep, prior knowledge, etc.) that your model doesn’t consider.

 R² acts as a measure of how well your independent variables are linked to your output variable, helping you understand their impact on the predictions made by your model.


```python
linreg=LinearRegression()
linreg.fit(X_train_scaled,y_train)
y_pred=linreg.predict(x_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean Absolute error ",mae)
print("R2 Score ",score)
plt.scatter(y_test,y_pred)
```

However, Lasso uses an L1 penalty (the absolute value of the coefficients) instead of the L2 penalty used by Ridge.

## Ridge 
* Ridge regression is a type of linear regression that includes a regularization term to prevent overfitting by penalizing large coefficients. The key difference from standard linear regression is that Ridge regression adds an 'L2 penalty (the square of the magnitude of coefficients)' to the loss function, which helps shrink the coefficients toward zero.

This regularization helps in cases where multicollinearity (high correlation between predictors) exists, or the model has many features, by reducing variance at the cost of a slight increase in bias. As a result, Ridge regression provides better generalization on unseen data compared to standard linear regression, especially when dealing with complex or noisy datasets.



<hr>

* L1 penalty (Lasso) forces some coefficients to become exactly zero, effectively performing feature selection. This means it can simplify the model by eliminating irrelevant features.
* L2 penalty (Ridge), on the other hand, shrinks coefficients but rarely forces them to zero, meaning all features remain in the model.

```python
from sklearn.linear_model import Ridge
```

```python
rid=Ridge()
rid.fit(X_train_scaled,y_train)
y_pred=rid.predict(x_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean Absolute error ",mae)
print("R2 Score ",score)
plt.scatter(y_test,y_pred)
```

# RidgeCV
* RidgeCV is a variant of Ridge regression that automatically tunes the regularization parameter λ (alpha) using cross-validation. Instead of manually specifying the regularization strength (as in standard Ridge regression), RidgeCV selects the optimal value of λ by testing multiple values and finding the one that minimizes the prediction error.

* How it works:
1. Cross-validation is used to split the training data into subsets, and the model is trained on some subsets and validated on others.
2. RidgeCV tries different values of λ (alphas) on these training subsets and evaluates the performance (error) on the validation sets.
3. It selects the λ that results in the best cross-validation performance (i.e., the smallest validation error).

```python
from sklearn.linear_model import RidgeCV
```

```python
ridcv=RidgeCV(cv=10)
ridcv.fit(X_train_scaled,y_train)
y_pred=ridcv.predict(x_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean Absolute error ",mae)
print("R2 Score ",score)
plt.scatter(y_test,y_pred)
```

## ElasticNet



* ElasticNet is a type of linear regression that combines the penalties of both Lasso (L1) and Ridge (L2) regression. It is useful when you have a dataset with many features, some of which may be highly correlated, or when you expect both feature selection (like Lasso) and shrinkage (like Ridge) to improve the model's performance.

```python
from sklearn.linear_model import ElasticNet
```

```python
en=ElasticNet()
en.fit(X_train_scaled,y_train)
y_pred=en.predict(x_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean Absolute error ",mae)
print("R2 Score ",score)
plt.scatter(y_test,y_pred)
```

```python
from sklearn.linear_model import ElasticNetCV
```

```python
encv=ElasticNetCV(cv=10)
encv.fit(X_train_scaled,y_train)
y_pred=encv.predict(x_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean Absolute error ",mae)
print("R2 Score ",score)
plt.scatter(y_test,y_pred)
```
